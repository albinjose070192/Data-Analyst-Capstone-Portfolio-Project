```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('The_Cancer_data_1500_V2.csv')
```

```python
df.head()
```

```python
 df.describe()
```

```python
df['Diagnosis'] = df['Diagnosis'].astype('str')
df['CancerHistory'] = df['CancerHistory'].astype('str')
df['Smoking'] = df['Smoking'].astype('str')
df['Gender'] = df['Gender'].astype('str')
df['GeneticRisk'] = df['GeneticRisk'].astype('str')
```

```python
df.describe()
```

```python
print('The average age for positive cancer diagnosis is ' + str(df['Age'][df['Diagnosis']=='1'].mean().round(0)) + ' years')
```

```python
print('The median age for positive cancer diagnosis is ' + str(df['Age'][df['Diagnosis']=='1'].median().round(0)) + ' years')
```

```python
print('Diagnosis of cancer is most prominent among individuals aged between ' + str(df['Age'][df['Diagnosis']=='1'].mean().round(0) - df['Age'][df['Diagnosis']=='1'].std().round(0)) + ' to ' + str(df['Age'][df['Diagnosis']=='1'].mean().round(0) + df['Age'][df['Diagnosis']=='1'].std().round(0)) + ' years')
```

My Observations:
1. In general, cancer diagnosis drastically increases with age. Especially post 50, the chances of positive diagnosis increases drastically. 
2. Although women generally don't get positively diagnosed with cancer before 50, the occurence of positive diagnosis in women before 50 is way higher than that for men.
3. As expected, alcohol intake and BMI are positively correlated with positive cancer diagnosis. Moreover, number of hours of physical activity is negatively correlated with a positive diagnosis. Therefore, it is strongly advised to stay active, stay away from alcohol, and maintain your weight to reduce the chances of getting diagnosed with cancer.
4. As a whole, women have more cases of positive diagnoses than men
5. Among men, there is a higher chance of recurrence if there is a past personal history of cancer diagnosis. Women are more likely to get diagnosed with cancer with no past personal history whatsoever.
6. Surprisingly, smoking plays a less important role in positive cancer diagnosis. This may be skewed by the fact that fewer percentage of women smokers get diagnosed with cancer as compared to men. Because of the absolute number of women that get diagnosed, the overall numbers show that smoking and genetic risk play a significantly lower role in positive cancer diagnosis, as opposed to contemporary knowledge. 
7. However, in case of smoking, the data is highly skewed where women generally tend to be less diagnosed with cancer if they smoke as compared to men. This may be because the type of cancers that women get diagnosed with more frequently may not be primarily caused due to smoking
8. The risk of getting diagnosed is alarmingly high among women with high genetic risk consuming over 3 units of alcohol per week or smoking regularly or both. Women with high genetic risk should definitely stay away from smoking or drinking alcohol 
9. Interestingly, men who smoke regularly and drink over 3 units of alcohol per week have a higher chance of getting diagnosed with cancer than women
10. A higher percentage of men are at the risk of getting diagnosed with cancer if they have a BMI higher than 25 and lack enough physical activity (minimum 2 hours per week). This risk is significantly reduced for higher BMI men if they put in atleast 2 hours of physical activity per week.
11. The risk of high BMI and low physical activity increases the risk of cancer in women drastically if they also consume alcohol or smoke or both regularly
